```python
import json
import os

files = os.listdir('.')
files = [f for f in files if f.endswith('.json')]
files.sort()
files
```

```python
res = []

for i, f in enumerate(files):
    with open(f) as fp:
        res.append(json.load(fp))
res
```

```python
json.dump(res, open('results.json', 'w'))
```
